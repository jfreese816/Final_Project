```python
from instagramy import InstagramUser
from instagramy import InstagramPost
import pandas as pd
```

```python
# Grab session id: 
# log into Instagram in browser, 
    # inspect page --> 
    # click on application tab --> 
    # expand cookies (left hand side under 'Storage') --> 
    # click on cookie for https://instagram.com --> 
    # copy session id from the table
    
session_id = "145470550%3Apk3dmP5fGaSHHO%3A6"
```

```python
# Separately, we created a list of Instagram users from whom we want to pull data
# Import list of users

# Read in separate list of pre-selected usernames into pandas dataframe
user_df = pd.read_csv('resources/top_ig_urls_MJ.csv', header=None)

# Show usernames
user_df
```

```python
# Creat empty lists
user_followers = []
user_posts = []
user_following = []
user_bio = []

for x in user_df[0]:
    user = InstagramUser( x, sessionid=session_id)

    # Scrape specific info
    user_followers.append( user.number_of_followers)
    user_posts.append( user.number_of_posts)
    user_following.append( user.number_of_followings)
    user_bio.append( user.biography)
```

```python
# Create dataframe for holding the information we scraped
user_list_df = pd.DataFrame( columns= ['account', 'bio','followers', 'following', 'posts'])
```

```python
# Add scraped info into the dataframe
user_list_df.account = user_df[0]
user_list_df.bio = user_bio
user_list_df.followers = user_followers
user_list_df.following = user_following
user_list_df.posts = user_posts

# Show dataframe
user_list_df
```

```python
user_list_df.to_csv('resources/top_ig_user_data_MJ.csv')
```
